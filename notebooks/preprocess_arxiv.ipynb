```
import scispacy
import spacy
from tqdm import tqdm
import pickle

import multiprocessing

the_queue = multiprocessing.Queue()
spacy.require_gpu()
import en_core_sci_sm
nlp = en_core_sci_sm.load()
import os


with open(os.path.join('.', 'arxiv_tuple_pickle'), 'rb') as f:
    dic = pickle.load(f)


processed = nlp.pipe(texts=dic, as_tuples=True, batch_size=8)


def to_file(doc):
    with open(os.path.join('./arxiv-processed', doc[1]), 'w') as ff:
        print(list(map(lambda x: x.string, list(doc[0].sents))), file=ff)


for doc in tqdm(processed, total=len(dic)):
    to_file(doc)
```
